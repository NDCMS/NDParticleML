```python
names = {}
names['cQei'] = 0
names['cQl3i'] = 1
names['cQlMi'] = 2
names['cbW'] = 3
names['cpQ3'] = 4
names['cpQM'] = 5
names['cpt'] = 6
names['cptb'] = 7
names['ctG'] = 8
names['ctW'] = 9
names['ctZ'] = 10
names['ctei'] = 11
names['ctlSi'] = 12
names['ctlTi'] = 13
names['ctli'] = 14
names['ctp'] = 15

WC_to_analyze_1 = ['cpQ3', 'cpQM', 'cpt', 'cptb', 'cQei', 'ctei', 'ctli', 'ctW'] # The first of the two WCs to graph
WC_to_analyze_2 = ['cbW', 'ctG', 'ctp', 'cQl3i', 'cQlMi', 'ctlTi', 'ctlSi', 'ctZ'] # The second of the two WCs to graph
```

```python
# Automatic damage checking for frozen
damaged = np.full((8),0)
for num in np.arange(8):
    WC1 = WC_to_analyze_1[num]
    WC2 = WC_to_analyze_2[num]
    try:
        loaded = np.load(f'likelihood_{WC1}_{WC2}.npz')
        for key in list(loaded.keys()):
            print (loaded[key])
    except:
        damaged[num] = 1

for num in np.arange(8):
    if (damaged[num]):
        WC1 = WC_to_analyze_1[num]
        WC2 = WC_to_analyze_2[num]
        print (f'likelihood_{WC1}_{WC2}.npz is damaged!')
```

```python
# Automatic damage checking for profiled
damaged = np.full((8),0)
for num in np.arange(8):
    WC1 = WC_to_analyze_1[num]
    WC2 = WC_to_analyze_2[num]
    try:
        loaded = np.load(f'likelihood_profiled_{WC1}_{WC2}.npz')
        for key in list(loaded.keys()):
            print (loaded[key])
    except:
        damaged[num] = 1

for num in np.arange(8):
    if (damaged[num]):
        WC1 = WC_to_analyze_1[num]
        WC2 = WC_to_analyze_2[num]
        print (f'likelihood_profiled_{WC1}_{WC2}.npz is damaged!')
```
